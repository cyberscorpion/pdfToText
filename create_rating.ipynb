```python
def readingFile(filename):
    f = open(filename,"r")
    data=[]
    c=0
    for row in f:
        if c==0:
            c=1
            continue
        r = row.split(',')
        r[len(r)-1] = r[len(r)-1][:-2]
        r[0] = r[0][1:]
        #print(r)
        #for i in range(len(r)):
            #print(i,'->',r[i],end=' ::: ')
        #print()
        r[0] = int(r[0])
        #print('r[0]=',r[0],"\t",type(r[0]))
        #print(r[1],"\t",type(r[1]))
        #r[2] = r[2][:-2]
        #print(r[2],"\t",type(r[2]))
        e = [ int(r[0]) , int(r[1]) , int(r[2])]
        data.append(e)
    f.close()
    return data


readingFile("aman_rating.csv")
```

```python
import csv
unique_skills = []
with open('unique_skills.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    for row in reader:
        print(row[0],type(row))
        unique_skills.append(row[0])

csvFile.close()
print("unique_skills = ",unique_skills,type(unique_skills[0]))
```
