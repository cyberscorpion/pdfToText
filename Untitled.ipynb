```python
import numpy as np
r = []
for index,row in df1.iterrows():
    row = row[~row.isnull()]
    row = [float(np_float) for np_float in row]
    
    for i in range(len(row)):
        print(i,row[i],type(row[i]))
        if float(row[i]) != np.nan:
            #print(str(int(row[i])),end=',')
            s = str(index+1)+','+str(int(row[i]))+',1'
            print(s)
            r.append(s)
    print()
```

```python
df2 = pd.DataFrame(r)
df2.to_csv("aman_rating.csv",index=False)
```

```python
def readingFile(filename):
    f = open(filename,"r")
    data=[]
    c=0
    for row in f:
        if c==0:
            c=1
            continue
        r = row.split(',')
        r[len(r)-1] = r[len(r)-1][:-2]
        r[0] = r[0][1:]
        #print(r)
        for i in range(len(r)):
            print(i,'->',r[i],end=' ::: ')
        print()
        r[0] = int(r[0])
        print('r[0]=',r[0],"\t",type(r[0]))
        #print(r[1],"\t",type(r[1]))
        #r[2] = r[2][:-2]
        #print(r[2],"\t",type(r[2]))
        e = [ int(r[0]) , int(r[1]) , int(r[2])]
        data.append(e)
    f.close()
    return data


readingFile("aman_rating.csv")
```
