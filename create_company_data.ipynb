```python
df = pd.read_csv("company.csv")
df
```

```python
from collections import defaultdict
company_id = defaultdict(int)
company_name = df['company']
company_name1 = list(company_name)
company_name = list(set(company_name1))
print(company_name1)
print()
print(company_name)
com_id=[]
#print(type(company_name),company_name)
cid=10
for i in company_name:
    company_id[i] = cid
    cid = cid+1
#print(company_id)
for c in company_name1:
    #print(c,company_id[c])
    com_id.append(company_id[c])
print(com_id)
com_id = pd.DataFrame(com_id)
com_id
df['com_id'] = com_id
```

```python
dfs = pd.read_csv("unique_skills.csv",header=None)
dfs.head()
#job_skill = list(df['skill'])
#for i in job_skill:
    #print(i)
dfs = list(dfs[0])
job_id = defaultdict(int)
for i in range(len(dfs)):
    job_id[dfs[i]]=i
job_id
job=[]
job_skill = list(df['skill'])
for i in range(len(job_skill)):
    j = job_skill[i].split(',')
    x=""
    for k in range(len(j)):
        x = x+str(job_id[j[k]])+','
    print(x)
    x=x.rstrip(',')
    job.append(x)
#print("job = ",job)
job = pd.DataFrame(job)
job.to_csv("aman.csv",index=False,header=None)
```
